```python
import pandas as pd

# ['Reaction_index', 'Substrate', 'Product', 'Activator', 'Inhibitor']
l_reaction = []

s_twoway = set([
    'mMTCH', 'cMTCH', 'mMTD', 'cMTD', 'SAHH', 'mSHMT', 'cSHMT',
    'b_GLY_c', 'b_SER_c', 'b_MET_c', 'mSERc', 'mGLYc', 'mHCOOHc',
    'mFTS',
])

for i, reaction in enumerate(mod.Nmatrix.getLabels()[1]):
    reaction = reaction.lstrip('V_')
    l_sub, l_pro = [], []

    for j, n in enumerate(mod.Nmatrix()[:, i]):
        metab = mod.Nmatrix.getLabels()[0][j]

        if n < 0:
            l_sub.append(metab)
        elif n > 0:
            l_pro.append(metab)

    if reaction == 'BHMT':
        l_reaction.append([reaction, ' '.join(l_sub), ' '.join(l_pro), '', 'sam'])
    elif reaction == 'CBS':
        l_reaction.append([reaction, ' '.join(l_sub), ' '.join(l_pro), 'sam sah', ''])
    elif reaction == 'GNMT':
        l_reaction.append([reaction, ' '.join(l_sub), ' '.join(l_pro), '', 'sah'])
    elif reaction == 'MATI':
        l_reaction.append([reaction, ' '.join(l_sub), ' '.join(l_pro), '', 'sam'])
    elif reaction == 'MATIII':
        l_reaction.append([reaction, ' '.join(l_sub), ' '.join(l_pro), 'sam', ''])
    elif reaction == 'MTHFR':
        l_reaction.append([reaction, ' '.join(l_sub), ' '.join(l_pro), '', 'sam'])

    else:
        l_reaction.append([reaction, ' '.join(l_sub), ' '.join(l_pro), '', ''])

        if reaction in s_twoway:
            l_reaction.append([reaction, ' '.join(l_pro), ' '.join(l_sub), '', ''])

df = pd.DataFrame(l_reaction)
df = df.rename(columns={0: 'Reaction', 1: 'Substrate', 2: 'Product', 3: 'Activator', 4: 'Inhibitor'})
df
```

```python
import pandas as pd

# ['Reaction_index', 'Substrate', 'Product', 'Activator', 'Inhibitor']
l_reaction = []

s_twoway = set([
    'mMTCH', 'cMTCH', 'mMTD', 'cMTD', 'SAHH', 'mSHMT', 'cSHMT',
    'b_GLY_c', 'b_SER_c', 'b_MET_c', 'mSERc', 'mGLYc', 'mHCOOHc',
    'mFTS',
])

for i, reaction in enumerate(mod.Nmatrix.getLabels()[1]):
    reaction = reaction.lstrip('V_')
    l_sub, l_pro = [], []

    for j, n in enumerate(mod.Nmatrix()[:, i]):
        metab = mod.Nmatrix.getLabels()[0][j]

        if n < 0:
            l_sub.append(metab)
        elif n > 0:
            l_pro.append(metab)

    if reaction == 'BHMT':
        l_reaction.append([reaction, ' '.join(l_sub), ' '.join(l_pro), '', 'sam'])
    elif reaction == 'CBS':
        l_reaction.append([reaction, ' '.join(l_sub), ' '.join(l_pro), 'sam sah', ''])
    elif reaction == 'GNMT':
        l_reaction.append([reaction, ' '.join(l_sub), ' '.join(l_pro), '', 'sah'])
    elif reaction == 'MATI':
        l_reaction.append([reaction, ' '.join(l_sub), ' '.join(l_pro), '', 'sam'])
    elif reaction == 'MATIII':
        l_reaction.append([reaction, ' '.join(l_sub), ' '.join(l_pro), 'sam', ''])
    elif reaction == 'MTHFR':
        l_reaction.append([reaction, ' '.join(l_sub), ' '.join(l_pro), '', 'sam'])

    elif reaction in s_twoway:
        l_reaction.append([reaction + '_1', ' '.join(l_sub), ' '.join(l_pro), '', ''])
        l_reaction.append([reaction + '_2', ' '.join(l_pro), ' '.join(l_sub), '', ''])

    else:
        l_reaction.append([reaction, ' '.join(l_sub), ' '.join(l_pro), '', ''])

df = pd.DataFrame(l_reaction)
df = df.rename(columns={0: 'Reaction', 1: 'Substrate', 2: 'Product', 3: 'Activator', 4: 'Inhibitor'})
df
```

```python
import numpy as np
from ibuffpy import ReactionNetwork


df = []
for r in l_reaction:
    l = []
    for i in r:
        i = i if i != '' else np.nan
        l.append(i)
    df.append(l)

net = ReactionNetwork.from_pandas(
    pd.DataFrame(df), ker_basis='svd', cq_basis='rref', sep=' '
)
```

```python
print(f'The number of chemicals is {net.M}.')
print(f'The number of reactions is {net.R}.')
print(f'The set of chemicals is {net.cpd_list_noout}.')
print(f'The set of reactions is {net.reactionNames}.')
```

```python
print(f'Cycles \n{net.ns}')
print(f'Conserved quantities\n{net.ns2}')
```

```python
# sensitivity matrix

import numpy as np

pd.set_option('display.max_columns', 100)

sens_mat = net.compute_smat_mean(N=10)
#sens_mat = pd.DataFrame(sens_mat[:net.M, :-2], columns=net.reactionNames, index=net.cpd_list_noout)
#sens_mat = pd.DataFrame(sens_mat[:net.M, 24:-2], columns=net.reactionNames[24:], index=net.cpd_list_noout)
d = np.concatenate((sens_mat[:12, 25:-2], sens_mat[19:net.M, 25:-2]))
sens_mat = pd.DataFrame(d, columns=net.reactionNames[25:], index=net.cpd_list_noout[:12]+net.cpd_list_noout[19:])
sens_mat.applymap(lambda x: 0 if np.abs(x) < 1e-10 else 1)
```

```python
# sensitivity matrix (sign)

sens_sign = net.compute_smat_sign()
#pd.DataFrame(sens_sign[:net.M, :-2], columns=net.reactionNames, index=net.cpd_list_noout)
#pd.DataFrame(sens_sign[:net.M, 24:-2], columns=net.reactionNames[24:], index=net.cpd_list_noout)
d = np.concatenate((sens_sign[:12, 25:-2], sens_sign[19:net.M, 25:-2]))
pd.DataFrame(d, columns=net.reactionNames[25:], index=net.cpd_list_noout[:12]+net.cpd_list_noout[19:])
```
